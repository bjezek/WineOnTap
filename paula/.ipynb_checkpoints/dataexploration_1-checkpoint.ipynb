```python
import pandas as pd
from pathlib import Path
import hvplot.pandas
```

```python
df = pd.read_csv(Path("winemag-data-130k-v2.csv"))
df2 = pd.read_csv(Path("wine-quality.csv"))
df.head()
```

```python
df2.head()
```

```python
X = df[["points", "price"]]
```

```python
df.describe()
```

```python
#df.groupby("province").count()
```

```python
# Start by importing the K-means algorithm
from sklearn.cluster import KMeans
```

```python
# Create and initialize the K-means model instance for 2 clusters
model = KMeans(n_clusters=10, random_state=1)
model2 = KMeans(n_clusters=10, random_state=1)

# Print the model
#model
```

```python
# Fit the data to the instance of the model
X = X.dropna()
df2 = df2.dropna()
model.fit(X)
model2.fit(df2)
```

```python
# Make predictions about the data clusters using the trained model
wine_clusters = model.predict(X)
quality_clusters = model2.predict(df2)

# Print the predictions
#print(customer_ratings)
```

```python
# Create a copy of the DataFrame
X_copy = X.copy()
df2_copy = df2.copy()

# Add a column to the DataFrame that contains the customer_ratings information
X_copy['cluster'] = wine_clusters
df2_copy['cluster'] = quality_clusters

# Review the DataFrame
#service_rating_predictions_df.head()
```

```python
# Plot the data points based on the customer rating
X_copy.hvplot.scatter(
    x="points", 
    y="price", 
    by="cluster"
)
```

```python
# Plot the data points based on the customer rating
df2_copy.hvplot.scatter(
    x="sulphates", 
    y="alcohol", 
    by="cluster"
)
```
